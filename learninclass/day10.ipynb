```python
number = [1,2,3,4]
# number[0] *= 5
# print(number)

#list comprehension
number = [num * 2 for num in number]
print(number)

#for loop
for num in number :
    num * 2
    print(num)

#for loop
list = []
for index, num in enumerate(number) : #จับคู่ระหว่าง Index กับ value ข้างใน
    number[index] = num * 2
    list.append(num)

print(list)

#for loop
for index in range(len(number)):
    number[index] *= 2

print(number)
```
