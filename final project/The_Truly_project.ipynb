```python
def linear_regression(series_x, series_y):

    n_bars = len(series_x)

    sum_x = sum(series_x)
    print(sum_x)

    sum_y = sum(series_y)
    print(sum_y)

    sum_xy = 0
    for index in range(n_bars):
        sum_xy += series_x[index] * series_y[index]
    print(sum_xy)

    sum_square_x = 0
    for index in range(n_bars):
        sum_square_x += series_x[index] ** 2
    print(sum_square_x)

    sum_square_y = 0
    for index in range(n_bars):
        sum_square_y += series_y[index] ** 2
    print(sum_square_y)
    
# The solution form cramer'rule to find slope and interception

    slope_of_linear = ((n_bars * sum_xy) - (sum_x * sum_y)) / \
        ((n_bars * sum_square_x) - (sum_x ** 2))

    intercept_of_linear = ((sum_y * sum_square_x) - (sum_x * sum_xy)) / \
        ((n_bars * sum_square_x) - (sum_x ** 2))

# The formular that tell correlation coefficient(r) that model in range -1,1 (x,y have linear correlation)
# near 1, -1 that mean x,y have linear correlation
# if near 0 mean x, y doesn't has linear correlation

    correlation_coefficient = ((n_bars * sum_xy) - (sum_x * sum_y)) / \
        ((n_bars * sum_square_x - sum_x ** 2) *
         (n_bars * sum_square_y - sum_y ** 2)) ** 0.5

    return slope_of_linear, intercept_of_linear, correlation_coefficient


series_x = index
series_y = list(data["close"])

slope, intercept, correlation_coefficient = linear_regression(
    series_x, series_y)

print("Slope :", slope)
print("Intercept :", intercept)
print("correlation coefficient :", correlation_coefficient)
```

```python
trend_line = [slope * x + intercept for x in series_x]

# graph
plt.figure(figsize=(10, 5))
plt.plot(series_x, series_y, label='Actual Close Price', color='blue', marker='o')
plt.plot(series_x, trend_line, label='Linear Trend Line', color='red')
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.title('Stock Price vs Linear Regression Trend')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
x_date = input(f"YYYY-MM-DD for exemple: 2025-11-06")  # input in format YYYY-MM-DD
print(x_date)

x_date = pd._libs.tslibs.timestamps.Timestamp(x_date + ' 20:30:00') #เพราะว่าไม่อยากให้เป็น -1 เพราะมันนับถอยหลัง ถ้าต้องการข้อมูลของวันที่ 10 ก็จะส่งก่อนหน้าเข้ามา

print((x_date - data.index[0]).days) #ลบข้างในคือระยะห่างวัน ได้ออกมาเป็นเลข  1 2 3 ... สารมารถเอาไปใช้ได้

def predict(another_day):

    y = (slope * another_day) + intercept
    return y

predict_prize = predict((x_date - data.index[0]).days)

print(predict_prize)
```
