## Nested Function

```python
def outer():
    def inner():
        print("I am inner function")
    inner()  # เรียกใช้งานภายใน outer

outer()
# inner()  # ถ้าเรียกข้างนอก จะ Error
```

```python
def outer():
    print("I am inner function")
outer()
```

```python
def outer(number):
    def square(x):
        return x ** 2
    
    result = [square(n) for n in number]
    print(result)

outer([1, 2, 3])
```

```python
def outer(number):
    def square(x):
        return x ** 2
    
    result = []

    for member in number : # member ที่เป็นค่าใน number 
        square_value = square(member) 
        # ตัวแปล square_value นี้คือหลังจากที่เราวน 1 รอบมันจะเก็บผลลัพธืจาก square(member) -> ซึ่งก็คือ inner function (ให้มันยกกำลังสอง) เสร็จเเล้วก็จะไปต่อ
    
        result.append(square_value) #เอาตัวแปล square_value ที่มีค่าในนั้นเก็บไว้ที่ list วนไปเรื่อย ๆ

    print(result) # ปริ้นออกมาหลังจบ loop เเต่ยังไม่เเสดงเพราะว่ายังไม่เรียกใช้ outer()
    # square()
outer([1,2,3]) #outer ไม่ได้ return result
# print(outer([1,2,3]))
```

```python
#อยากปริ้นเป็นเเค่ตัวเลขก็เเค่

def outer(number):
    def square(x):
        return x ** 2
    
    result = []

    for member in number :
        result.append(square(member))

    
    for value in result: #เพราะว่า ตัวแปลชื่อ value วนใน result เเล้วปริ้นค่าออกมาทีละตัว จนครบจำนวนใน list
        print(value)
    
outer([1,2,3]) 
```

## lambda Fucntion

```python
square = lambda x : x ** 2 #คือแบบ input : output
print(square(5))
```

```python
def square(x):
    return x ** 2
print(square(5))
```

```python
numbers = [1, 2, 3, 4]

result = list(map(lambda x: x ** 2, numbers) )
# (map(lambda x: x ** 2, numbers)) คือ ส่งค่าเข้าไปใน lambda function 
# list เพื่อทำให้คำตอบมันเป็น List (ถ้าไม่อยากให้ัมันเป็น list ก็เอาออกได้)

print(result)
```

```python
numbers = [1, 2, 3, 4]
result = []

for x in numbers:
    result.append(x ** 2)

#หรือไม่ใช้ map, list ครอย ก็จะออกมาทีละตัวเลข
for member in result:
    print(member)

#ถ้าไม่มีวนทีละตัวใน result ก็จะออกมาเป็น List เพระาเราให้ result = []
print(result)
```
