```python
def linear_regression(series_x : list, series_y : list) -> tuple[float, float, float]:

    n_bars = len(series_x)
# The data points are available in the dataset, means the number of stock closing prices collected over a certain period
    
    sum_x = sum(series_x)
    # print(sum_x)

    sum_y = sum(series_y)
    # print(sum_y)

    sum_xy = 0
    for index in range(n_bars):
        sum_xy += series_x[index] * series_y[index]
    # print(sum_xy)

    sum_square_x = 0
    for index in range(n_bars):
        sum_square_x += series_x[index] ** 2
    # print(sum_square_x)

    sum_square_y = 0
    for index in range(n_bars):
        sum_square_y += series_y[index] ** 2
    # print(sum_square_y)
    

    slope_of_linear = ((n_bars * sum_xy) - (sum_x * sum_y)) / \
        ((n_bars * sum_square_x) - (sum_x ** 2))

    intercept_of_linear = ((sum_y * sum_square_x) - (sum_x * sum_xy)) / \
        ((n_bars * sum_square_x) - (sum_x ** 2))

# The solution form cramer'rule to find slope and interception

    correlation_coefficient = ((n_bars * sum_xy) - (sum_x * sum_y)) / \
        ((n_bars * sum_square_x - sum_x ** 2) *
         (n_bars * sum_square_y - sum_y ** 2)) ** 0.5
    
# The formular that tell correlation coefficient(r) that model in range -1,1
# near 1, -1 that mean x,y have linear correlation (1 = strong positive correlation, -1 = strong negative correlation)
# if near 0 mean x, y doesn't has linear correlation = weak or no correlation

    return slope_of_linear, intercept_of_linear, correlation_coefficient


series_x = index
series_y = list(data["close"])

slope, intercept, correlation_coefficient = linear_regression(
    series_x, series_y)

print("Slope :", slope)
print("Intercept :", intercept)
print("correlation coefficient :", correlation_coefficient)
```

```python
trend_line = [slope * x + intercept for x in series_x]

# graph
plt.figure(figsize=(10, 5))
plt.plot(series_x, series_y, label='Actual Close Price', color='blue', marker='o')
plt.plot(series_x, trend_line, label='Linear Trend Line', color='red')
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.title('Stock Price vs Linear Regression Trend')
plt.legend()
plt.grid(True)
plt.show()
```
