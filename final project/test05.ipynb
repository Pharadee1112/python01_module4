#### ได้ time series อยุ่ในรูปของ linear regression

```python
index = [0]

for date in data.index[1:]:
    value = date - data.index[0]
    index.append(value.days)

print(index)
print(len(index))
print(type(index))
```

```python
# print(data["close"])
print(type(data["close"]))
```

```python
def linear_regression(series_x, series_y):

    n_bars = len(series_x)

    sum_x = sum(series_x)
    sum_y = sum(series_y)

    sum_xy = 0
    for index in range(n_bars):
        sum_xy += series_x[index] * series_y[index]
    
    sum_square_x = 0
    for index in range(n_bars):
        sum_square_x += series_x[index] ** 2

    slope_of_linear = (n_bars * sum_xy - sum_x * sum_y) / (n_bars * sum_square_x - sum_x**2)
    intercept_of_linear = (sum_y * sum_square_x - sum_x * sum_xy) / (n_bars * sum_square_x - sum_x**2)

    return slope_of_linear, intercept_of_linear

series_x = index
series_y = list(data["close"])

slope, intercept = linear_regression(series_x, series_y)
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
import matplotlib.pyplot as plt
```

```python
trend_line = [slope * x + intercept for x in series_x]

# วาดกราฟ
plt.figure(figsize=(10, 5))
plt.plot(series_x, series_y, label='Actual Close Price', color='blue', marker='o')
plt.plot(series_x, trend_line, label='Linear Trend Line', color='red')
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.title('Stock Price vs Linear Regression Trend')
plt.legend()
plt.grid(True)
plt.show()
```
