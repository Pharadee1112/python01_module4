## modifying list

```python
number = [1, 2, 3] #วิธีเพิ่ม 4 เข้าไปใน list

#1.
number.append(4) #in=place relpacement operation = add the number to the list
print(number)

#2.
number += [4] #อันนี้ก็สามารถเพิ่มตัวอักษรเข้าไปได้ ใน list
print(number)

#3.
number.insert(3, 4) #ช้ามาก
print(number)

#4.
number.extend([4])
print(number)
```

crate a list from 1-100

remove the odd number that is not divisible by 3

final list -> [1, 2, 4,.....]

count many number left

```python
number = list(range(1, 101))
print(number)


for num in number:
    if num % 3 == 0 and num % 2 != 0 :
        number.remove(num)

print(number)
print(len(number))   
```

```python
number = []
for num in range(1, 101):
    if num % 2 != 0 and num % 3 == 0 :
        continue
    number.append(num)

print(number)
```

```python
number = []
for num in range(1, 101):
    if not (num % 2 == 0 and num % 3 != 0) :
        continue
    number.append(num)

print(number)
```

```python
number = [num for num in range (1, 101) if \
          (num % 2 == 0 or num % 3 != 0)]
print(number)
```

```python
#list comprehension
number = [element for element in range(5)]
print(number)

#Tradition
number = []
for element in range(5):
    number.append(element)

print(number)
```
